```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Carga del Dataset**

```python
dataset = pd.read_csv('/content/movie_metadata.csv')
```

```python
dataset.info()
print('\n')
print ('El número de filas y columnas del Dataframe es')
dataset.describe()
```

# **Búsqueda de duplicados**

```python
duplicados= dataset.duplicated().sum()
print('Hay', duplicados , 'duplicados')
print('\n')
```

# **Eliminar los duplicados**



```python
dataset_cleaned = dataset.drop_duplicates()
print('La información del Dataset eliminada es:')
dataset_cleaned.info()
print('\n')
print('fueron eliminados', duplicados, 'registros duplicados')
print(f'Duplicados eliminados. Filas restantes {dataset_cleaned.shape[0]}')
```

**Identificación de Valores Nulos:**

```python
print('Valores faltantes por columna:')
print('\n')
dataset_cleaned.isnull().sum()

```

# **Visualización de Variables:**

```python
dataset_cleaned['imdb_score'].hist(bins=20)
plt.title('Número de peliculas IMDB')
plt.xlabel('Puntaje IMDB')
plt.ylabel('Cantidad de peliculas')
plt.grid(True)
plt.show()


```

# **Diagrama de Barras**

```python
dataset_cleaned['genres'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Géneros de Peliculas')
plt.xlabel('Generos')
plt.ylabel('Cantidad de peliculas')
plt.grid(True)
plt.show()
```

# **Analisis de Correlación**

```python
dataset_cleaned['budget'] = dataset_cleaned['budget'].str.replace('$', '')

#Correlación budget
correlación = dataset_cleaned[['budget', 'gross', 'imdb_score']].corr()

#Mapa de color
sns.heatmap(correlación, annot=True)
plt.title('Correlación entre Gross, Budget y IMDB Score')
plt.show()
```

```python
# 7. Gráficos de Dispersión:

# Crea un gráfico de dispersión para analizar la relación entre budget y gross.
dataset_reduc=dataset_cleaned.head(50)
dataset_reduc.plot(kind='scatter', x='budget', y='gross', alpha=0.5)
plt.title('Budget vs Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.grid(True)
plt.show()
```

```python
# Haz un gráfico de dispersión entre imdb_score y budget.
dataset_reduc.plot(kind='scatter', x='imdb_score', y='budget', alpha=0.5)
plt.title('IMDB vs budget')
plt.xlabel('imdb')
plt.ylabel('budget')
plt.grid(True)
plt.show()
```
